```python
import pandas 
import yfinance
```

```python
import statsmodels.api as sm
```

```python
import datetime

# Define tickers and date range (YTD)
tickers = ["SOXX", "SPY"]
today = datetime.date.today()
start = datetime.date(today.year, 1, 1)
end = today

# Download data
df = yfinance.download(tickers, start=start, end=end, progress=False)["Adj Close"]

# Compute daily returns
returns = df.pct_change().dropna()

# Run regression: SOXX return ~ SPY return (add constant for intercept)
X = sm.add_constant(returns["SPY"])
y = returns["SOXX"]

model = sm.OLS(y, X)
results = model.fit()

beta = results.params["SPY"]
print(f"YTD Beta loading of SOXX vs SPY: {beta:.3f}")
```
